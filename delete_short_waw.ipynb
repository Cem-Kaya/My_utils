```python
import pandas as pd
import os
from pydub import AudioSegment

# Load the CSV files
train_df = pd.read_csv('metadata_train.csv', sep='|')
eval_df = pd.read_csv('metadata_eval.csv', sep='|')

# Define the minimum duration threshold in milliseconds (e.g., 1000 ms = 1 second)
min_duration_ms = 4000

# Function to get the duration of a WAV file
def get_wav_duration(file_path):
    audio = AudioSegment.from_wav(file_path)
    return len(audio)

# List to store files to be deleted
files_to_delete = []

# Check durations and mark files for deletion in the training set
for index, row in train_df.iterrows():
    wav_path = row['audio_file']
    full_path = os.path.join(wav_path)
    if get_wav_duration(full_path) < min_duration_ms:
        files_to_delete.append(wav_path)
        train_df.drop(index, inplace=True)

# Check durations and mark files for deletion in the evaluation set
for index, row in eval_df.iterrows():
    wav_path = row['audio_file']
    full_path = os.path.join(wav_path)
    if get_wav_duration(full_path) < min_duration_ms:
        files_to_delete.append(wav_path)
        eval_df.drop(index, inplace=True)

# Delete marked files from the filesystem
for file in files_to_delete:
    os.remove(file)

# Save the updated CSV files
train_df.to_csv('updated_metadata_train.csv', sep='|', index=False)
eval_df.to_csv('updated_metadata_eval.csv', sep='|', index=False)

# Output the list of deleted files
print("Deleted files:", files_to_delete)
```
